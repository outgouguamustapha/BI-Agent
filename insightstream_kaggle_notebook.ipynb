### 💡 The Core Problem
Traditional Generative AI data analysis runs blindly. Business intelligence requires extreme numbers precision, strict data confidentiality, and mathematical reproducibility. If an AI "hallucinates" a decimal or modifies a database row, it poses an existential enterprise risk.

### 🧠 The InsightStream Solution
**InsightStream** is an autonomous, self-contained multi-agent pipeline that transforms high-level business queries into boardroom-ready intelligence. By separating responsibilities through the **DIG (Description, Introspection, Goal-Setting) Framework**, it isolates data discovery from quantitative logic, operating behind an unbreakable, engine-level security sandbox.


## 🏗️ System & Security Architecture

InsightStream coordinates a sequential team of three specialized AI agents, mapped directly to real-world data team roles, enforcing absolute validation at every handoff.

### 🔄 The DIG Framework Sequence Flow

```text

              ▼
   ┌─────────────────────────────┐
```

1. **Product Manager Agent (BI Lead)** ➔ *Goal-Setting*: Translates the high-level plain English query into structured analytical milestones and defines target performance constraints.
2. **Data Engineer Agent (SQL Guard)** ➔ *Description & Introspection*: Inspects column metadata, maps primary relational constraints, and converts the PM's strategy into highly optimized, read-only SQL SELECT queries.
3. **Data Analyst Agent (Reproducibility Lead)** ➔ *Quantitative Matrix Execution*: Ingests the data frame payloads, computes business-critical KPIs (MoM MRR, user trajectory velocities, API volume anomalies), and exports the finalized business portfolio assets.


## 🛡️ Defense-in-Depth Security Perimeter

To prove this architecture is safe for corporate transactional ecosystems, InsightStream enforces a **two-layer database sandbox wrapper** that physically prevents data mutations:

* **Layer 1: Prior-to-Connection Static Filter:** The query is stripped of all comment indicators and scanned across rigid word boundaries (`\b...\b`) to immediately intercept forbidden write operations (`INSERT`, `UPDATE`, `DROP`, `DELETE`, `PRAGMA`).
* **Layer 2: Engine-Level Connection Authorizer:** As an airtight defense-in-depth policy, the pipeline registers a compilation callback using `sqlite3.Connection.set_authorizer`. If a prompt injection bypasses the static layer, the core database engine blocks the transaction bytecode at compile-time, triggering a safety exception.

> **Zero-Mutation Guarantee:** InsightStream is mathematically incapable of deleting, dropping, or altering a single row of company records.
